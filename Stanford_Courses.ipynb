```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://explorecourses.stanford.edu/'
search_url = 'https://explorecourses.stanford.edu/print?q={}&descriptions=on&filter-term-Winter=off&academicYear=&filter-term-Summer=off&filter-term-Autumn=off&filter-departmentcode-{}=on&filter-term-Spring=off&page=0&filter-coursestatus-Active=on&catalog='
html = requests.get(url).text
bs = BeautifulSoup(html, 'lxml')
```

```python
searches = [url+link['href'] for link in bs.findAll('a') if link['href'].startswith('search')]
```

```python
tokens = [link.get_text().strip().split(' ')[-1][1:-1] for link in bs.findAll('a') if link['href'].startswith('search')]
```

```python
courses = []
for token in tokens:
    html = requests.get(search_url.format(token, token)).text
    bs = BeautifulSoup(html, 'lxml')
    
    for elem in bs.findAll('div', {'class':'searchResult'}):
        courseNumber = elem.find('span', {'class':'courseNumber'}).get_text().replace(':', '')
        courseTitle = elem.find('span', {'class':'courseTitle'}).get_text()
        courseDescription = elem.find('div', {'class':'courseDescription'}).get_text()
        courseAttributes = ' '.join(elem.find('div', {'class':'courseAttributes'}).get_text().replace('\r', '').replace('\n', '').replace('\t', '').split(' '))
        courses.append([courseNumber, courseTitle, courseDescription, courseAttributes.strip().replace('    ', '')])
```

```python
df = pd.DataFrame(courses, columns=['Number', 'Title', 'Description', 'Attributes'])
df['Token'] = df['Number'].str.split(' ').str[0]
df['Number'] = df['Number'].str.split(' ').str[1]
df.head()
```

```python
search_terms = ['visualization', 'tensorflow', 'deep learning', 'convolutional neural network', 'reinforcement learning', 'machine learning', 'graph theory']
```

```python
treffer = []
for word in search_terms:
    for elem in df.values:
        if word in elem[1].lower() or word in elem[2].lower():
            treffer.append(df[(df['Number']==elem[0]) & (df['Token']==elem[4])]['Description'].values[0])
treffer = set(treffer)
len(treffer)
```

```python
for text in treffer:
    number, token, title, description = df[df['Description']==text][['Number', 'Token', 'Title', 'Description']].values[0]
    print(token, number, title)
    print(description)
    print()
```
