# Download the TU Dortmund course catalog for a better search

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://www.lsf.tu-dortmund.de/qisserver/rds?state=wtree&search=1&trex=step&root120181=72353&P.vx=kurz'
html = requests.get(url).text
bs = BeautifulSoup(html, 'lxml')

links = [[link['href'], False] for link in bs.findAll('a', {'class':'ueb'})]
df = pd.DataFrame(links, columns=['Link', 'Checked'])
```

```python
events = []
while not df[df['Checked']==False].empty:
    
    for i, link in enumerate(df[df['Checked']==False]['Link']):
        df.loc[df[df['Link']==link].index, 'Checked'] = True
        bs = BeautifulSoup(requests.get(link).text, 'lxml')
        links = [[link['href'], False] for link in bs.findAll('a', {'class':'ueb'}) if link['href'] not in df['Link']]

        for table in bs.findAll('table', {'summary':'Übersicht über alle Veranstaltungen'}):
            for tr in table.findAll('tr'):
                try:
                    number = tr.findAll('td')[0].get_text().strip()
                    title = tr.findAll('td')[1].get_text().strip().split('\n')[0].replace('\r', '')
                    events.append([number, title])
                except:
                    pass
        
        temp_df = pd.DataFrame(links, columns=('Link', 'Checked'))
        df = pd.concat([df, temp_df]).reset_index(drop=True).drop_duplicates()
    
        if i%100==0:
            print('Links-All: {},\tLinks-Checked: {},\tEvents: {}'.format(df.shape[0], df[df['Checked']==True].shape[0], len(events)))
```

```python
events_df = pd.DataFrame(events, columns=['Number', 'Title'])
events_df = events_df.drop_duplicates()
events_df.to_csv('TU_Dortmund_Vorlesungen.csv', index=False)
```
